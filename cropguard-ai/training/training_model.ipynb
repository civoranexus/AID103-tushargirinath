```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
```

```python

# -----------------------------
# Configuration
# -----------------------------
IMAGE_SIZE = 224
NUM_CLASSES = 5
MODEL_SAVE_PATH = "../models/cropguard_model.h5"

```

```python
# -----------------------------
# Dummy Dataset Loader
# -----------------------------
def load_dummy_data(num_samples=50):
    X = np.random.rand(num_samples, IMAGE_SIZE, IMAGE_SIZE, 3)
    y = np.random.randint(0, NUM_CLASSES, size=(num_samples,))
    y = to_categorical(y, NUM_CLASSES)
    return X, y
```

```python
# -----------------------------
# Build CNN Model
# -----------------------------
def build_model():
    model = Sequential([
        Conv2D(16, (3,3), activation="relu", input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)),
        MaxPooling2D(2,2),
        Conv2D(32, (3,3), activation="relu"),
        MaxPooling2D(2,2),
        Flatten(),
        Dense(64, activation="relu"),
        Dense(NUM_CLASSES, activation="softmax")
    ])

    model.compile(
        optimizer="adam",
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

```python
# -----------------------------
# Training Pipeline
# -----------------------------
def train():
    print("Loading dataset...")
    X, y = load_dummy_data()

    print("Building CNN model...")
    model = build_model()

    print("Training model...")
    model.fit(X, y, epochs=3, batch_size=8)

    print("Saving model...")
    model.save(MODEL_SAVE_PATH)

    print("Training completed successfully!")

if __name__ == "__main__":
    train()
```
